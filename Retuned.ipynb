```python
X = new_df.iloc[:, 1:14]
y = new_df.iloc[:, 0]


#print(X)
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 17, test_size = 0.2)
```

```python
X_train_clean = X_train.dropna()
y_train_clean = y_train[X_train_clean.index] 
X_test_clean = X_test.dropna()
y_test_clean = y_test[X_test_clean.index]  

print("Rows in training data before dropping NaN:", X_train.shape)
print("Rows in test data before dropping NaN:", X_test.shape)

print("Rows in training data after dropping NaN:", X_train_clean.shape)
print("Rows in test data after dropping NaN:", X_test_clean.shape)
```

```python
from xgboost.callback import EarlyStopping
model_xgboost = xgboost.XGBClassifier(learning_rate = 0.1, 
                                      max_depth = 2, 
                                      n_estimators = 1000,
                                      subsample = 0.5,
                                      colsample_bytree = 0.5,
                                      eval_metric = 'auc',
                                      early_stopping_rounds = 50,
                                      verbosity = 1 ,   
                                      alpha= 10.0,     
                                      reg_lambda = 15.0 
                                    
                                    
                                    
                                     )
#eval_set = [(X_train, y_train),(X_test, y_test)]
eval_set = [(X_test, y_test)]
model_xgboost.fit(X_train,
                  y_train,
                  eval_set = eval_set,
                  verbose = True
                 )
```

```python
y_train_pred = model_xgboost.predict_proba(X_train)[:,1]
y_test_pred = model_xgboost.predict_proba(X_test)[:,1]
print("AUC Train: {:.4f}\nAUC test: {:.4f}".format(
    roc_auc_score(y_train, y_train_pred),
    roc_auc_score(y_test, y_test_pred)
))
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Now we manually perform cross-validation with early stopping
from sklearn.metrics import roc_auc_score

cv_scores = []

for train_idx, val_idx in cv.split(X_train, y_train):
    X_train_cv, X_val_cv = X_train.iloc[train_idx], X_train.iloc[val_idx]
    y_train_cv, y_val_cv = y_train.iloc[train_idx], y_train.iloc[val_idx]
    
    # Fit the model on the current training fold with early stopping
    model_xgboost.fit(X_train_cv, y_train_cv, 
                      eval_set=[(X_val_cv, y_val_cv)], 
                      verbose=False)
    
    # Predict on validation data
    y_val_pred = model_xgboost.predict_proba(X_val_cv)[:, 1]
    
    # Compute the AUC score for this fold
    auc_score = roc_auc_score(y_val_cv, y_val_pred)
    cv_scores.append(auc_score)

# Print the mean AUC score across all folds
print(f"AUC Cross-validation scores: {cv_scores}")
print(f"Average AUC: {np.mean(cv_scores)}")
```
